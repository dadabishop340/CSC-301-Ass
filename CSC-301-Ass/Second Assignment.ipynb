# 1. Using a recursion to get a sum of first ten numbers

```python
def sum_recursive(n):
    if n==1:
        return 1
    else:
        return n + sum_recursive(n-1)
result = sum_recursive(10)
print("Sum of first ten numbers:", result)
```

# 2. Write a fibonacci series of 8 term

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
n_terms = 8 
series = []

for i in range(n_terms):
    series.append(fibonacci(i))
print("Fibonacci series of 8 terms:", series)
```

 # 3. Insert the following values into a dynamic arrays of capacity 2

### values = 10, 20, 30, 40, 50

```python
class DynamicArray:
    def __init__(self, capacity=2):
        self.capacity = capacity
        self.size = 0
        self.array = [None] * capacity

    def insert(self, value):
        if self.size == self.capacity:
            self._resize()
        self.array[self.size] = value
        self.size += 1

    def _resize(self):
        self.capacity *= 2
        new_array = [None] * self.capacity
        for i in range(self.size):
            new_array[i] = self.array[i]
        self.array = new_array

    def display(self):
        return self.array[:self.size]

# Using the dynamic array
values = [10, 20, 30, 40, 50]
dyn_array = DynamicArray()

for val in values:
    dyn_array.insert(val)

print("Final dynamic array:", dyn_array.display())
print("Final capacity:", dyn_array.capacity)
```

# 4. Perform a linear search on an array with the values:
### values= 2, 5, 7, 10, 14, 20 target is 10.

```python
def linearSearch(arr, targetVal):
    for i in range(len(arr)):
        if arr[i] == targetVal:
            return i
    return -1
mylist = [2, 5, 7, 10, 14, 20]
x = 10

result = linearSearch(mylist, x)
if result != -1:
    print("Found at index", result)
else:
    print("Not found")
```

# 5. Place the binary search on 1,3,5,7,9,11,13 target is 9

```python
def binarySearch(arr, targetVal):
  left = 0
  right = len(arr) - 1

  while left <= right:
    mid = (left + right) // 2

    if arr[mid] == targetVal:
      return mid

    if arr[mid] < targetVal:
      left = mid + 1
    else:
      right = mid - 1

  return -1

mylist = [1, 3, 5, 7, 9, 11, 13]
x = 9

result = binarySearch(mylist, x)

if result != -1:
  print("Found at index", result)
else:
  print("Not found")
```

# * State the time and space complexity of both

# 1. Linear Search
## Time Complexity

### --Worst Case: O(n)
(Target is at the end or not present)

### --Best Case: O(1)
(Target is the first element)

Average Case: O(n)

## Space Complexity

O(1)
(Uses only a few variables, no extra memory)

# 2. Binary Search
## Time Complexity

### --Worst Case: O(log n)
(Each step divides the array in half)

### --Best Case: O(1)
(Target found at the first mid check)

Average Case: O(log n)

## Space Complexity

Iterative Version: O(1)
(No extra memory except a few variables)

Recursive Version: O(log n)
(Because each recursive call adds a stack frame)
